# 1. Описание данных

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('telecom_users.csv')
df.head()
```

```python
df.info()
```

```python
df.TotalCharges.replace(' ', 0, inplace=True)
df.TotalCharges = df.TotalCharges.apply(float)
df.describe()
```

```python
len(df['customerID'].unique()) == df.shape[0]
```

Пропущенных данных в нашей таблице нет, в предыдущей ячейке убедились, что все строки с ID пользователей являются уникальными, повторов нет.

Преобразуем текстовые, категориальные переменные в численные для дальнейшей работы.

```python
bin_codes = {'No': 0, 'Yes': 1}

df['gender'] = df['gender'].map( {'Female': 0, 'Male': 1} ).astype(int)
df['Partner'] = df['Partner'].map(bin_codes).astype(int)
df['Dependents'] = df['Dependents'].map(bin_codes).astype(int)
df['PhoneService'] = df['PhoneService'].map(bin_codes).astype(int)
df['PaperlessBilling'] = df['PaperlessBilling'].map(bin_codes).astype(int)
df['Churn'] = df['Churn'].map(bin_codes).astype(int)
```

```python
df.head()
```

```python
df['CategoricalTotalCharges'] = pd.cut(df['TotalCharges'], 10, labels=False)

internet_cat_codes = {'No internet service': 0, 'No': 0, 'Yes': 1}
internet_cat = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
                'TechSupport', 'StreamingTV', 'StreamingMovies']

df = df.merge(pd.get_dummies(df[['customerID','InternetService']], prefix='InternetService',
                             columns=['InternetService']), on='customerID')
df = df.drop(['InternetService_No', 'InternetService'], axis = 1)

for cat in internet_cat:
    df[cat] = df[cat].map( internet_cat_codes ).astype(int)

df['MultipleLines'] = df['MultipleLines'].map( {'No phone service': 0, 'No': 0, 'Yes': 1} ).astype(int)
df['Contract'] = df['Contract'].map( {'Month-to-month': 1, 'One year': 2, 'Two year': 3} ).astype(int)

PaymentMethod_mapping = {'Electronic check': 1, 'Mailed check': 2, 'Bank transfer (automatic)': 3, 'Credit card (automatic)': 4}
df['PaymentMethod'] = df['PaymentMethod'].map(PaymentMethod_mapping)
    
df.head()
```

Избавимся от неинформативных переменных, также исключим переменную общих расходов, т.к. мы преобразовали ее в категориальную с целью избавиться от чрезмерного влияния больших цифр в работе алгоритмов.

```python
drop_elements = ['Unnamed: 0', 'customerID', 'TotalCharges']
df = df.drop(drop_elements, axis = 1)
```

# 2. Исследование зависимостей и формулирование гипотез

```python
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns

Y = df['Churn']
X = df
x_train, x_test, y_train, y_test = train_test_split(X, Y, random_state=0, stratify=Y)
```

```python
colormap = plt.cm.RdBu
plt.figure(figsize=(25,20))
plt.title('Pearson Correlation of Features', y=1.05, size=15)
sns.heatmap(x_train.astype(float).corr(),linewidths=0.1,vmax=1.0, 
            square=True, cmap=colormap, linecolor='white', annot=True)
```

```python
sns.pairplot(x_train[['tenure', 'MonthlyCharges', 'CategoricalTotalCharges']], 
                  height=2, plot_kws=dict(s=20), corner=True, diag_kind="kde")
```

Матрица корреляций демонстрирует нам незначительное кол-во коррелирующих между собой переменных, исключим из них переменную CategoricalTotalCharges, т.к. вряд ли клиенты ведут счет совокупных расходов на услуги, к тому же эту переменной можно заменить периодом пользования услугами.

```python
drop_elements = ['CategoricalTotalCharges', 'Churn']
X = X.drop(drop_elements, axis = 1)

x_train, x_test, y_train, y_test = train_test_split(X, Y, random_state=0)
```

# 3. Построение моделей для прогнозирования оттока

Возьмем для работы следующие модели: GradientBoostingClassifier, RandomForestClassifier, SVC, LogisticRegression. 
Зададим наборы параметров моделей для поиска среди них наилучших, обучим модели и найдем наилучшие параметры.

```python
from sklearn.model_selection import (GridSearchCV,
                                     train_test_split,
                                     StratifiedKFold)

from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression

from xgboost import XGBClassifier


gbc_params = {'learning_rate': np.arange(0.1, 0.6, 0.1)} 

rfc_params = {'n_estimators': range(10, 100, 10),
              'min_samples_leaf': range(1, 5)}

svc_params = {'kernel': ['linear', 'rbf'],
              'C': np.arange(0.1, 1, 0.2)}

lr_params = {'C': np.arange(0.5, 1, 0.1)}

skf = StratifiedKFold(n_splits=5, random_state=17, shuffle=True)
```

```python
gbc = GridSearchCV(GradientBoostingClassifier(random_state=17), param_grid=gbc_params, cv=skf)
gbc.fit(x_train, y_train)
print(f'Лучшее сочетание параметров для GradientBoostingClassifier: {gbc.best_params_}')

rfc = GridSearchCV(RandomForestClassifier(random_state=17), param_grid=rfc_params, cv=skf)
rfc.fit(x_train, y_train)
print(f'Лучшее сочетание параметров для RandomForestClassifier: {rfc.best_params_}')

svc = GridSearchCV(SVC(random_state=17), param_grid=svc_params, cv=skf)
svc.fit(x_train, y_train)
print(f'Лучшее сочетание параметров для SVC: {svc.best_params_}')

lr = GridSearchCV(LogisticRegression(random_state=17, max_iter=1000), param_grid=lr_params, cv=skf)
lr.fit(x_train, y_train)
print(f'Лучшее сочетание параметров для LogisticRegression: {lr.best_params_}')
```

```python
from sklearn.metrics import accuracy_score

print(f'Accuracy for GradientBoostingClassifier: {accuracy_score(y_test, gbc.predict(x_test))}')
print(f'Accuracy for RandomForestClassifier: {accuracy_score(y_test, rfc.predict(x_test))}')
print(f'Accuracy for SVC: {accuracy_score(y_test, svc.predict(x_test))}')
print(f'Accuracy for LogisticRegression: {accuracy_score(y_test, lr.predict(x_test))}')
```

На основе обученных моделей построим стекинг.

```python
%%time
from sklearn.ensemble import StackingClassifier


models = [('GBC', gbc), ('RFC', rfc), ('SVC', svc), ('LR', lr)]

meta_clf = StackingClassifier(estimators=models)
meta_clf.fit(x_train, y_train)

print(f'Accuracy for meta-algorithm: {accuracy_score(y_test, meta_clf.predict(x_test))}')
```

Метрика точности для модели стекинга незначительно отличается от метрики для модели логистической регрессии; учитывая значительные временные затраты работы мета-алгоритма и возможность интерпретировать параметры для регрессии, в качестве прогнозной модели предпочтительнее выглядит модель лог. регрессии.

Попробуем применить к нашим данным снижение размерности, построим pipeline - нормализуем данные, найдем наилучшее значение числа новых переменных и наилучшие наборы параметров для моделей. Обучим модели.

```python
from sklearn.decomposition import PCA
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

pca_params = {'n_components': np.arange(x_train.shape[1], 1, -1)}

gbc_pca = make_pipeline(StandardScaler(), GridSearchCV(PCA(random_state=17), param_grid=pca_params, cv=skf), GridSearchCV(GradientBoostingClassifier(random_state=17), param_grid=gbc_params, cv=skf))
gbc_pca.fit(x_train, y_train)

rfc_pca = make_pipeline(StandardScaler(), GridSearchCV(PCA(random_state=17), param_grid=pca_params, cv=skf), GridSearchCV(RandomForestClassifier(random_state=17), param_grid=rfc_params, cv=skf))
rfc_pca.fit(x_train, y_train)

svc_pca = make_pipeline(StandardScaler(), GridSearchCV(PCA(random_state=17), param_grid=pca_params, cv=skf), GridSearchCV(SVC(random_state=17), param_grid=svc_params, cv=skf))
svc_pca.fit(x_train, y_train)

lr_pca = make_pipeline(StandardScaler(), GridSearchCV(PCA(random_state=17), param_grid=pca_params, cv=skf), GridSearchCV(LogisticRegression(random_state=17), param_grid=lr_params, cv=skf))
lr_pca.fit(x_train, y_train)
```

```python
print(f'Accuracy for GradientBoostingClassifier: {accuracy_score(y_test, gbc_pca.predict(x_test))}')
print(f'Accuracy for RandomForestClassifier: {accuracy_score(y_test, rfc_pca.predict(x_test))}')
print(f'Accuracy for SVC: {accuracy_score(y_test, svc_pca.predict(x_test))}')
print(f'Accuracy for LogisticRegression: {accuracy_score(y_test, lr_pca.predict(x_test))}')
```

```python
print(f'Оптимальное значение размерности для GradientBoostingClassifier: {gbc_pca["gridsearchcv-1"].best_params_}')
print(f'Оптимальное значение размерности для RandomForestClassifier: {rfc_pca["gridsearchcv-1"].best_params_}')
print(f'Оптимальное значение размерности для SVC: {svc_pca["gridsearchcv-1"].best_params_}')
print(f'Оптимальное значение размерности для LogisticRegression: {lr_pca["gridsearchcv-1"].best_params_}')
```

# 4. Сравнение качества моделей 

Подводя итоги можно сделать вывод о том, снижение размерности не дает эффекта. Таким образом, для прогнозирования оттока клиентов предпочтительно использовать логистическую регрессию с параметром регуляризации С = 0.79, что на тестовой выборке обеспечивает точность порядка 80 %.

```python
lr.best_estimator_.coef_[0]

print('Отношение вероятности ухода клиента при наличии фактора (переменная = 1) / при росте фактора на единицу:\n')

for ind, i in enumerate(lr.best_estimator_.coef_[0]):
    print(f'Переменная {x_train.columns[ind]}, отношение: {round(np.exp(1)**i, 2)}')
```

Можно попробовать интерпретировать полученные оценки коэффициентов в лог. регрессии, так, например, для мужчин вероятность оттока на прибл. 10 % ниже, чем для женщин, для пенсионеров же вероятность ухода выше почти на 25 %. 

Исходя из оценок коэффициентов можно было бы сделать следующие выводы:
1) повысить лояльность и доступность услуг для пенсионеров;
2) стараться вовлечь клиентов через услуги телефонной связи, - это очень мощный фактор удержания;
3) продолжать поддерживать функции онлайнбэкапа, тех. поддержки, онлайн безопасности возможно за счет ресурсов, направленных на поддержание стриминговых сервисов;
4) не стоит обманываться низким значением коэф-та при переменной мес. расходов, т.к. мы не категоризировали эту переменную, она осталась в абсолютном выражении.
